```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = '/content/test.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Exploratory Data Analysis (EDA)

plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Ratings vs Main Category
plt.figure(figsize=(14, 8))
sns.boxplot(x='maincateg', y='Rating', data=df)
plt.title('Boxplot of Ratings by Main Category')
plt.xlabel('Main Category')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Sorting the dataframe by 'Rating' in descending order and selecting the top 30 products
top_30_products = df.sort_values(by='Rating', ascending=False).head(30)

# Creating a horizontal bar plot
plt.figure(figsize=(12, 10))
sns.barplot(x='Rating', y='title', data=top_30_products, palette='viridis')
plt.title('Top 30 Products with Highest Ratings')
plt.xlabel('Rating')
plt.ylabel('Product Title')
plt.show()
```

```python

df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

df_sorted = df.sort_values(by='Rating', ascending=False)

# Top 30 products
top_30 = df_sorted.head(30)

# Below 30 products
below_30 = df_sorted.tail(30)
```

```python
# Creating scatter plot for top 30 products
plt.figure(figsize=(14, 8))
sns.scatterplot(x='title', y='Rating', data=top_30)
plt.title('Scatter Plot for Top 30 Products')
plt.xlabel('Product Title')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Creating scatter plot for below 30 products
plt.figure(figsize=(14, 8))
sns.scatterplot(x='title', y='Rating', data=below_30)
plt.title('Scatter Plot for Below 30 Products')
plt.xlabel('Product Title')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Assuming 'actprice1' column represents the offers
df['actprice1'] = pd.to_numeric(df['actprice1'], errors='coerce')

# Sorting the DataFrame by 'actprice1' in descending order
df_sorted = df.sort_values(by='actprice1', ascending=False)

# Top 25 products with the highest offers
top_25_offers = df_sorted.head(25)

# Bottom 25 products with the lowest offers
bottom_25_offers = df_sorted.tail(25)
```

```python
# Creating bar chart for top 25 products
plt.figure(figsize=(14, 8))
sns.barplot(x='title', y='actprice1', data=top_25_offers, palette='viridis')
plt.title('Top 25 Products with Highest Offers')
plt.xlabel('Product Title')
plt.ylabel('Offer Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Creating bar chart for bottom 25 products
plt.figure(figsize=(14, 8))
sns.barplot(x='title', y='actprice1', data=bottom_25_offers, palette='magma')
plt.title('Bottom 25 Products with Lowest Offers')
plt.xlabel('Product Title')
plt.ylabel('Offer Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Creating a scatter plot with linear regression line
plt.figure(figsize=(10, 8))
sns.lmplot(x='actprice1', y='Rating', data=df, height=6, aspect=2, scatter_kws={'s': 30}, line_kws={'color': 'red'})
plt.title('Scatter Plot with Linear Regression Line')
plt.xlabel('Actual Price')
plt.ylabel('Rating')
plt.show()
```

```python
# Ploting the correlation matrix using a heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
